```python
### با عرض سلام و ادب خدمت شما ، ببخشید من کد رو توی محیط گوگل کولب مجبور شدم بنویسم اگر شما کد رو خواستید ران کنید و بررسی کنید 
### هر جا که cv2_imshow 
## نوشته شده به کد زیر تغغیر دهید 
##  cv2.imshow('winname' , img )
## البته cv2.waitkey فراموش نشه :) 

from mtcnn.mtcnn import MTCNN 
import cv2 
from google.colab.patches import cv2_imshow 
```

```python
img = cv2.imread('/content/download (1).jfif') 
```

```python
cv2_imshow(img)
```

```python
detector = MTCNN()
faces    = detector.detect_faces(img)
for i in faces : 
  print(i) 
```

```python
for i in faces : 
  cv2.rectangle(img ,  (i['box'][0] - 2  , i['box'][1]-2)  , (i['box'][0]+i['box'][2]+2 , i['box'][1]+i['box'][3]+2)  , 
                (0,255,0) , 2 )
```
